```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import re
pat = re.compile(r"#(\w+)")
```

```python
def create_hashtags_file(filename):
    df = pd.read_csv(filename)
    result = []
    for index, row in df.iterrows():
        hashtags = pat.findall(row["tweet_text"])
        result.extend(hashtags)
    return ", ".join(result)
```

```python
def create_wordcloud_img(text):
    wordcloud = WordCloud(max_font_size=50, max_words=50,collocations=False,width=300, height=300).generate(text)
    plt.figure(figsize=(10,10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
hashtags = create_hashtags_file("dataset/brexit-keyword-06-11-19-data.csv")
```

```python
create_wordcloud_img(hashtags)
```
